## import & load data

```python
import seaborn as sns
%matplotlib inline
```

```python
df = sns.load_dataset('tips')
```

```python
df.info()
```

# **Distribution Plot**

## distplot

```python
sns.distplot(df['total_bill'])
```

```python
sns.distplot(df['total_bill'],kde=False,bins=30)
```

## jointplot

matchup 2 distplot for bivariate data (e.g : scatter, hex, reg)

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind='scatter')
```

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind='hex')
```

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind='reg')
```

## pairplot

hubungan antara satu variabel antar yang lain (numerical) dan mendukung color hue untuk categorical column

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='sex',palette='viridis')
```

## rugplot

menampilkan dash mark untuk setiap point dalam univariate distribution

```python
sns.rugplot(df['total_bill'])
```

# Grids

## pairgrid

grid adalah tipe plot umum yang memungkinkan untuk mapping tipe plot dalam baris dan kolom

```python
import matplotlib.pyplot as plt
```

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
iris.info()
```

```python
sns.PairGrid(iris)
```

```python
g = sns.PairGrid(iris)
g.map(plt.scatter)
```

```python
# Map to upper,lower, and diagonal
g = sns.PairGrid(iris)
g.map_diag(plt.hist)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

## facetgrid

```python
df.head()
```

```python
sns.FacetGrid(df,col='time',row='smoker')
```

```python
g = sns.FacetGrid(df, col="time",  row="smoker")
g = g.map(plt.hist, "total_bill")
```

```python
g = sns.FacetGrid(df, col="time",  row="smoker",hue='sex')
# Notice hwo the arguments come after plt.scatter call
g = g.map(plt.scatter, "total_bill", "tip").add_legend()
```
