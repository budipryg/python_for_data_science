```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(0,5,6)
y = x**2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y,color='r',linestyle='--',linewidth=2,marker='o',markerfacecolor='g',markeredgecolor='black')
plt.xlabel('X label')
plt.ylabel('Y label')
plt.title('Simple Plot')
plt.xlim(0,5)
plt.ylim(0,25)
plt.show()
```

```python
fig = plt.figure()

ax = fig.add_axes([0.1,0.1,1,0.8])

ax.plot(x,y,'r')
ax.set_xlabel('X label')
ax.set_ylabel('Y label')
ax.set_title('Object Oriented Plot')

plt.show()
```

```python
fig = plt.figure(figsize=(12,8))

ax1 = fig.add_axes([0.1,0.1,1,0.9])
ax2 = fig.add_axes([0.2,0.5,0.4,0.4])

ax1.plot(x,y,'r')
ax1.set_xlabel('X label')
ax1.set_ylabel('Y label')
ax1.set_title('Object Oriented Plot')

ax2.plot(y,x,'b')
ax2.set_xlabel('X label')
ax2.set_ylabel('Y label')
ax2.set_title('Small Plot')

plt.show()
```

```python
fig, ax = plt.subplots(1,2)
```

```python
ax
```

```python
for a in ax:
    a.plot(x,y,'g')
    a.set_xlabel('X')
    a.set_ylabel('Y')
    a.set_title('Title')

fig
```

```python
fig, ax = plt.subplots(1,2,figsize=(12,8))

for a in ax:
    a.plot(x,y)
    a.set_xlabel('X')
    a.set_ylabel('Y')
    a.set_title('Title')

plt.tight_layout()
```
